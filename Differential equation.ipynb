```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline 

# function that returns dy/dt
def model(T,t):
    k = 0.054
    dTdt = -k*(T - 5) 
    return dTdt

# initial condition
T0 = 100

# time points
t = np.linspace(0,200)

# solve ODE
T_sol = odeint(model,T0,t)

x1, y1 = [5, 200], [5, 5]

# plot results
plt.plot(t, T_sol, 'b-', linewidth=2.0)
plt.plot(x1, y1,'r--', linewidth=2.0)
plt.xlabel("Time (min)")
plt.ylabel("temperature of the soup T (C)")
plt.legend(["N","Ta"])
plt.show()
```
