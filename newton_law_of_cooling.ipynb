```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline 

# function that returns dy/dt
def model(T,t):
    global T_s
    k = 0.054
    dTdt = -k*(T - T_s) 
    return dTdt

# initial condition
T0 = int(input("Enter the intial temp of your object: "))

# time points
t = np.linspace(0,200)

# temp of the surrounding
T_s = int(input("Enter the temp of the surrounding: "))

# solve ODE
T_sol = odeint(model,T0,t)

# Line for the temp of the surrounding 
x1, y1 = [5, 200], [5, 5]

# 20 degrees 
x2, y2 = [0, 200], [20, 20]


# plot results
plt.plot(t, T_sol, 'b-', linewidth=2.0)
plt.plot(x1, y1,'r--', linewidth=2.0)
plt.plot(x2, y2,'g--', linewidth=2.0)
plt.xlabel("Time (min)")
plt.ylabel("temperature of the soup T (C)")
plt.legend(["N","Ta", "20C"])
plt.show()
```
